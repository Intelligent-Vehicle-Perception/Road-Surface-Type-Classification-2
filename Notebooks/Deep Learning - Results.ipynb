<h5>Execution Logs</h5>

```
def getExecutionLogs():

    execution_logs = {}
    columns = ['window', 'train_loss',	'val_loss',	'train_acc', 'val_acc']
    models = os.listdir(os.path.join(experiments_folder))
    models.sort(reverse=True)

    for model in models:
        
        for placement in ["Below Suspension", "Above Suspension", "Dashboard"]:

            placement_logs = []
            
            for experiment in range(1, 4):

                file = os.path.join(experiments_folder, model, placement, "Experiment " + str(experiment), "experiment-execution-log.csv")

                if os.path.isfile(file):
                    content = pd.read_csv(file)
                    placement_logs.append(content[columns])

            execution_logs[model + " " + placement] = placement_logs    
    
    return execution_logs
```

```
def printValuesDnn(include=None, exclude=None, allResults=True, summarizedResults=False, showExperiments=False):

    execution_logs = getExecutionLogs()

    for model in execution_logs.keys():
        
        experiments = len(execution_logs[model])
        sum = None
        
        for i in range(0, experiments):

            if sum is None:
                sum = execution_logs[model][i].copy()
            else:
                sum += execution_logs[model][i].copy()
        
        if experiments > 0:

            if include is not None:

                show = False

                for part in include:
                    if part in model:
                        show=True
                        break

            elif exclude is not None:

                show = True

                for part in exclude:
                    if part in model:
                        show=False
                        break

            else:
                show = True

            if show:

                pd.set_option("float_format", '{:0.5f}'.format)
                print("Model " + model + (" | Experiments " + str(experiments) if experiments < 3 else ""))
                result = sum.copy()/experiments

                if allResults:
                    display(result)

                if summarizedResults:
                    pd.set_option("float_format", '{:0.2f}%'.format)
                    result = result[['window','val_acc']]
                    result["val_acc"] *= 100
                    result = result.transpose()
                    result["average"] = result.mean(axis=1) 
                    display(result)

                print("\n")

                if showExperiments:
                    
                    for i in range(0, experiments):
                        print("Model " + model + " | Experiment " + str(i+1))
                        display(execution_logs[model][i])
                        print("\n")
```

```
# printValuesDnn(include=["LSTM"], allResults=False, summarizedResults=True)
printValuesDnn(include=["LSTM"])
```

```
# printValuesDnn(include=["GRU"], allResults=False, summarizedResults=True)
printValuesDnn(include=["GRU"])
```

```
# printValuesDnn(include=["CNN"], allResults=False, summarizedResults=True)
printValuesDnn(include=["CNN"], allResults=True)
```

```
# printValuesDnn(include=["CNN"], allResults=False, summarizedResults=True)
printValuesDnn(include=["CNN Time Domain 7 Features Dashboard"], allResults=True, showExperiments=True)
```
